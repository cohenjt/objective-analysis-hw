```python
import numpy as np
from numpy.linalg import eig
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
```

```python
data = xr.open_dataset('../midterm2021.nc').data
print(data)
```

# 1. Hovmoller Plot

```python
sample = data.sel(samples=slice(0,200))

fig = plt.figure(figsize=(7, 14))
ax = fig.add_subplot()
ax.set_aspect(1/2)

plt.title('Hovmoller Plot')
plt.xlabel('Structure')
plt.ylabel('Samples')

cf = plt.contourf(sample.structure, sample.samples, sample.values, cmap='bwr')
plt.contour(sample.structure, sample.samples, sample.values, colors='k', linewidths=0.5)

cbar = plt.colorbar(cf, fraction=0.09, pad=0.025)

plt.show()
```

# 2. Eigenvalue Spectrum

```python
# Remove mean and divide by standard deviation
demeaned = data - data.mean(dim='samples')
std_data = demeaned / data.std(dim='samples') # Use this if we want normalized data

norm = 0
if norm:
    analyze = std_data
else:
    analyze = demeaned

# Calculate covariance matrix and perform eigenanalysis
cov = np.cov(analyze.T, ddof=0)
d, e = eig(cov)
```

```python
nstruct = len(analyze.structure)
nsamp = len(analyze.samples)
auto_corrs = xr.DataArray(
    data=np.zeros((nstruct, 2*nsamp-1)),
    dims=('structure', 'lags'),
    coords={'structure': np.linspace(start=0, stop=nstruct-1, num=nstruct),
            'lags': np.linspace(-nsamp+1, nsamp-1, num=2*nsamp-1)})

def auto_corr(x):
    xx = (x - np.mean(x)) / np.std(x)
    rhos = np.correlate(xx, xx, mode='full') / len(x)
    return rhos

for i, s in enumerate(auto_corrs.structure):
    auto_corrs.loc[{'structure': s}] = auto_corr(analyze.loc[{'structure': i}])
    
autocorrs1 = auto_corrs.sel(lags=slice(0, nsamp))  # just use positive lags
```

```python
# Calculate autocorrelation at lag 1 (average of structures' mean and max values)
amean = abs(autocorrs1).mean(dim='structure').isel(lags=1).values
amax = abs(autocorrs1).max(dim='structure').isel(lags=1).values

ralph = (amean + amax)/2
nstar = nsamp*(1-ralph*ralph)/(1+ralph*ralph)
north_fac = np.sqrt(2/nstar)
```

```python
L = d / sum(d)
LE = north_fac * L
xs = np.arange(0, len(L))
fig = plt.figure(figsize=(12, 8))
plt.errorbar(xs, L, yerr=LE, ecolor='k', capsize=5, lw=2, barsabove=True)

plt.title('Eigenvalue Spectrum')
plt.xlabel('Index')
plt.ylabel('Fraction of Variance Explained')

plt.show()
```

It looks like the first three principle components explain the vast majority of the variance, and the first five are linearly independent within error. Beyond the first five the eigenvalues overlap, so only the first five eigenvectors are significant according to the raw North criterion.

# 3. Eigenvectors

```python
fig = plt.figure(figsize=(12, 8))

plt.title('First Five Eigenvectors')
plt.xlabel('Structure')
plt.ylabel('Normalized EOF Amplitude')

nplt = 5
for i in range(nplt):
    frac = 100 * (d[i] / sum(d))
    plt.plot(e[:, i], label=f'EOF {i+1}: {frac:0.0f}% variance explained')

plt.grid()
plt.legend()
plt.show()
```

# 4. Principle Components

```python
# Project eigenvectors onto original data
z = xr.DataArray(
    data = np.matmul(e.T, data.values.T),
    dims = ('eigenvector', 'samples'))
z_samp = z.sel(samples=slice(0, 200-1))

fig = plt.figure(figsize=(12, 8))

plt.title('First Three Principle Components')
plt.xlabel('Samples')
plt.ylabel('PC Amplitude')

nplts = 3
for i in range(nplts):
    plt.plot(z_samp.samples, z_samp.isel(eigenvector=i), label=f'Principle Component {i+1}')

plt.legend()
plt.show()
```

# 5. Regressions

```python
# Calculate D from unstandardized data so we can check our regression method

Deof = np.matmul(e, np.sqrt(np.diag(d)))

# Test DD' = C and D'D = Lambda
c = np.matmul(Deof, Deof.T)
lam = np.matmul(Deof.T, Deof)
np.testing.assert_allclose(c, cov, atol=1e-12)
np.testing.assert_allclose(lam, np.diag(d), atol=1e-12)
```

```python
# Normalize principle components to have zero mean and unit variance
zm = (z - z.mean(dim='samples')) / (z.std(dim='samples'))
xm = demeaned.T

# Test ZZ' = IN
np.testing.assert_allclose(np.matmul(zm.values, zm.values.T)/nsamp, np.eye(nstruct, nstruct), atol=1e-10)

EM = np.matmul(xm.values, zm.values.T) / nsamp

# Test EM=D
np.testing.assert_allclose(Deof, EM, atol=1e-13)
```

```python
fig = plt.figure(figsize=(12, 8))

plt.title('Regressions of First Five Normalized PCs on Unnormalized Data')
plt.xlabel('Structure')
plt.ylabel('Amplitude in Physical Units')

nplts = 5
for i in range(nplts):
    plt.plot(EM[:, i], label=f'D {i+1}')

plt.legend()
plt.show()
```

# 6. Discussion

From the EOF analysis conducted here, we can conclude that a few structures likely exist in this data. From our initial glance at the Hovmoller plot, we can tell that the data is correlated in both space and time. Our further analysis shows that the main mode present in the data strongly resembles a cosine wave and explains about 40% of the overall variance. The second mode (20% of the variance) seems to represent an overall trend, and the third mode (14% of the variance) resembles a sine wave. Other significant modes are faster varying sine/cosine waves. Each of these modes may represent a distinct physical process, but because we have no a priori understanding of the data we are analyzing, we should be cautious of our overall interpretation.
